```python
import scanpy as sc
import pandas as pd

DATA_DIR = "../heart/raw_data/"

sc_ann_df = pd.read_csv(DATA_DIR+"Filtered/all_cells_meta_data_filtered.tsv", sep='\t', index_col=0)
sc_ann_df = sc_ann_df.dropna(subset=['nUMI'])

adata_sc = sc.read_csv(
    DATA_DIR+"filtered/all_cells_count_matrix_filtered.tsv",  # the directory with the `.mtx` file
    delimiter='\t',
    first_column_names=True                # use gene symbols for the variable names (variables-axis index)
    ).T

adata_sc.obs = sc_ann_df
adata_sc
```
